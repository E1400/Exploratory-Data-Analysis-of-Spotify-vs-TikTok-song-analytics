This exploratory data analysis (EDA) project investigates the impact of song features on popularity ranking between TikTok and Spotify. We were interested in this comparison to understand if the TikTok algorithm pushes certain songs to be more popular, compared to Spotify, which is a global streaming site. # Column descriptions

These are the columns in which we selected for both the Spotify and TikTok dataset as we believe that they will be important for our findings.

Danceability: describes how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity. A value of 0.0 is least danceable and 1.0 is most danceable.

Energy: a measure from 0.0 to 1.0 and represents a perceptual measure of intensity and activity. Typically, energetic tracks feel fast, loud, and noisy. For example, death metal has high energy, while a Bach prelude scores low on the scale. Perceptual features contributing to this attribute include dynamic range, perceived loudness, timbre, onset rate, and general entropy.

Key: The estimated overall key of the track. Integers map to pitches using standard Pitch Class notation . E.g. 0 = C, 1 = C♯/D♭, 2 = D, and so on. If no key was detected, the value is -1.

Mode: Mode indicates the modality (major or minor) of a track, the type of scale from which its melodic content is (0) major and (1) is minor. 

Loudness: The overall loudness of a track in decibels (dB). Loudness values are averaged across the entire track and are useful for comparing relative loudness of tracks. Loudness is the quality of a sound that is the primary psychological correlate of physical strength (amplitude). Values typical range between -60 and 0 db.

Speechiness: Speechiness detects the presence of spoken words in a track. The more exclusively speech-like the recording (e.g. talk show, audio book, poetry), the closer to 1.0 the attribute value. Values above 0.66 describe tracks that are probably made entirely of spoken words. Values between 0.33 and 0.66 describe tracks that may contain both music and speech, either in sections or layered, including such cases as rap music. Values below 0.33 most likely represent music and other non-speech-like tracks.

Acousticness: A confidence measure from 0.0 to 1.0 of whether the track is acoustic. 1.0 represents high confidence the track is acoustic.

Instrumentalness: Predicts whether a track contains no vocals. "Ooh" and "aah" sounds are treated as instrumental in this context. Rap or spoken word tracks are clearly "vocal". The closer the instrumentalness value is to 1.0, the greater likelihood the track contains no vocal content. Values above 0.5 are intended to represent instrumental tracks, but confidence is higher as the value approaches 1.0.

Liveness: Detects the presence of an audience in the recording. Higher liveness values represent an increased probability that the track was performed live. A value above 0.8 provides strong likelihood that the track is live.

Valence: A measure from 0.0 to 1.0 describing the musical positiveness conveyed by a track. Tracks with high valence sound more positive (e.g. happy, cheerful, euphoric), while tracks with low valence sound more negative (e.g. sad, depressed, angry).

Tempo: The overall estimated tempo of a track in beats per minute (BPM). In musical terminology, tempo is the speed or pace of a given piece and derives directly from the average beat duration.

Duration: Duration of song in milliseconds

## Why Did We Choose These Features?
These features were chosen to because they capture various aspects of the song that are known to influence its appeal and popularity among listeners:

Danceability: This feature describes how suitable a track is for dancing. Danceability is a key factor in determining the appeal of a song, especially in dance clubs and party environments.

Energy: Energy is a measure of the intensity and activity of a track. High-energy songs are often more engaging and enjoyable for listeners, which can contribute to their popularity.

Key: The key of a track can influence its emotional impact and how it resonates with listeners. Different keys can evoke different moods and feelings.

Mode: The selection of mode as a feature for our analysis stems from its significance in music theory and its potential influence on the perceived mood and emotional response of listeners. In music theory, mode refers to the arrangement of pitches within a scale, which can affect the overall tonality and character of a piece of music.

Loudness: Loudness is a measure of the overall volume of a track. Louder tracks can be more attention-grabbing and impactful, potentially contributing to their popularity.

Speechiness: Speechiness indicates the presence of spoken words in a track. This feature can be important in distinguishing between different genres and styles of music, which can affect the appeal of a song to different audiences.

Acousticness: Acousticness measures the degree to which a track is acoustic. Acoustic tracks often have a more intimate and organic feel, which can appeal to certain listeners and contribute to a song's popularity.

Instrumentalness: Instrumentalness predicts whether a track contains vocals or is purely instrumental. Instrumental tracks can have a wide appeal, especially in contexts where vocals may be distracting or undesirable.

Liveness: Liveness detects the presence of an audience in a recording, indicating whether a track was performed live. Live recordings can have a raw and authentic quality that resonates with listeners.

Tempo: Tempo is the speed or pace of a track, measured in beats per minute. The tempo of a song can influence its mood and how it is perceived by listeners.

Duration: The duration of a song can impact its appeal and popularity. Shorter songs may be more accessible and easier to consume, while longer songs can offer a more immersive listening experience.

```python
# import
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

## Popular Spotify Songs EDA

```python
# dataset 1
ss1 = pd.read_csv('spotify_top_charts_22.csv',encoding='latin1')
ss1.columns = [col.replace('_', ' ') for col in ss1.columns]
ss1.columns = [col.replace('%', ' ') for col in ss1.columns]
target_columns1 = ['artist names', 'track name', 'peak rank','danceability', 'energy', 'key', 
                   'loudness','mode','speechiness','acousticness','instrumentalness','liveness','tempo',
                   'duration ms']

cs1 = ss1[target_columns1]
cs1 = cs1.rename(columns={'peak rank':'popularity'})
cs1 = cs1[cs1['popularity'] <= 100]
cs1 = cs1.sort_values(by='popularity', ascending = True)
cs1['popularity'] = cs1['popularity'].apply(lambda x: abs(x - 100))
# rows organized in descending order by popularity (highest to lowest)
cs1 = cs1.sort_values(by='popularity', ascending = False)

#duration changed to seconds
cs1['duration ms'] = cs1['duration ms']/1000
cs1 = cs1.rename(columns={'duration ms':'duration sec'})

# removing duplicate row values/NaN
cs1 = cs1.drop_duplicates(subset=['track name'])
cs1 = cs1.drop_duplicates(subset=['popularity'])
cs1 = cs1.dropna()
cs1.head()
```

## TikTok songs 2022 EDA

```python
ts = pd.read_csv('TikTok_songs_2022.csv')
ts.columns = [col.replace('_', ' ') for col in ts.columns]

tiktok_target_columns = ['track name', 'artist name', 'track pop',
       'danceability', 'energy', 'loudness', 'mode', 'key', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration ms']

cts = ts[tiktok_target_columns]

cts = cts.rename(columns={'track pop': 'popularity'})

#duration changed to seconds
cts['duration ms'] = cts['duration ms']/1000
cts = cts.rename(columns={'duration ms':'duration sec'})

#dropping duplicate track names
cts = cts.drop_duplicates(subset=['track name'])
#dropping duplicates of popularity. One song is chosen at random as a representative of that rank in popularity
cts = cts.drop_duplicates(subset=['popularity'])
cts = cts.dropna()
cts = cts.sort_values(by='popularity', ascending = False)
cts.head()
```

Since we have a lot of observations for both the TikTok and Spotify Dataset, we will simply only take the top 50 songs to use in our exploratory data analysis.

```python
top_50_tiktok = cts[:50]
top_50_tiktok.head(5)
```

```python
top_50_spotify = cs1[:50]
top_50_spotify.head(5)
```

# TikTok (Tempo - Popularity)

```python
# Plotting the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='tempo', y='popularity', data=top_50_tiktok)
plt.title('TikTok Song Popularity vs. Tempo')
plt.xlabel('tempo')
plt.ylabel('popularity (TikTok songs)')
plt.grid(True)

# Fitting a regression line
X = top_50_tiktok['tempo'].values.reshape(-1, 1)
y = top_50_tiktok['popularity'].values.reshape(-1, 1)
reg = LinearRegression()
reg.fit(X, y)
plt.plot(X, reg.predict(X), color='red')

plt.show()

# Compute the correlation coefficient
correlation_coefficient = top_50_tiktok['tempo'].corr(top_50_tiktok['popularity'])
print("Correlation coefficient:", correlation_coefficient)

# Interpret the correlation coefficient
if correlation_coefficient > 0:
    print("There is a positive relationship between song rank and tempo.")
elif correlation_coefficient < 0:
    print("There is a negative relationship between song rank and tempo.")
else:
    print("There is no linear relationship between song rank and tempo.")
```

There is not a strong correlation between TikTok song's tempo and its popularity, linear regression line shows a slightly negative trend. The distribution range seems to be between about 60bpm to 190bpm beats per minute

## Spotify Key and Mode

```python
# Grouping the data by key and mode, and calculating the mean song popularity for each combination
mean_popularity_by_key_mode = top_50_spotify.groupby(['key', 'mode'])['popularity'].mean().sort_values(ascending=False)

# Plotting the bar plot
plt.figure(figsize=(12, 6))
mean_popularity_by_key_mode.unstack().plot(kind='bar', stacked=True)
plt.title("Spotify's Average Song Popularity by Key and Mode")
plt.xlabel('Musical Key')
plt.ylabel('Average Song Popularity (Spotify Charts)')
plt.xticks(rotation=45)
plt.legend(title='Mode', labels=['Major', 'Minor'])
plt.grid(axis='y')
plt.show()

# Grouping the data by key and mode, and calculating the mean song popularity for each combination
mean_popularity_by_key_mode = top_50_spotify.groupby(['key', 'mode'])['popularity'].mean().sort_values(ascending=False)

print("Mean Song Popularity by Key and Mode:")
print(mean_popularity_by_key_mode)
```
